```python
from matplotlib.font_manager as fm

rc('font', family ='Malgun Gothic')
font1 = {'size': 18, 'color': 'green'}

x_data = ['철수', '유리', '짱구']
y_data = [90, 85, 40]

plt.plot(x_data, y_data)
plt.title('국어 성적', fontdict=font1)

plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```
